# if elif vs 딕셔너리 속도 비교.

결론부터 말하자면 딕셔너리가 빠름. (불확실)<br>

코드의 실행 시간을 줄이기 위한 최적화를 고려해보겠습니다.<br>
현재 코드에서는 큐에 150개의 데이터가 삽입되고 처리되는데, 이 데이터 처리에 따른 시간은 큐에 데이터를 추가하는 데 걸리는 시간과 큐에서 데이터를 가져오고 처리하는 데 걸리는 시간으로 나뉩니다.<br>

데이터를 큐에 추가하는 부분은 최적화할 여지가 제한적이지만, 데이터 처리 부분은 최적화 가능합니다.<br>
코드에서는 큐에서 데이터를 가져올 때마다 딕셔너리에서 처리 함수를 찾아 호출하고 있습니다.<br>
이 작업은 반복적으로 수행되므로, 딕셔너리 조회와 함수 호출을 줄이는 방법으로 실행 시간을 줄일 수 있습니다.<br>

```python
import time
import queue

start_time = time.time()

# 큐 생성
my_queue = queue.Queue()

# 큐에 데이터 추가
for _ in range(50):
    my_queue.put("data1")
for _ in range(50):
    my_queue.put("data2")
for _ in range(50):
    my_queue.put("data3")
# 이곳에 수십 개의 데이터 추가

# 큐에서 데이터 분류 및 처리
while not my_queue.empty():
    data = my_queue.get()
    
    # 분류 조건에 따라 처리
    if data == "data1":
        # data1에 대한 처리 코드
        print("data1 처리 완료")
    elif data == "data2":
        # data2에 대한 처리 코드
        print("data2 처리 완료")
    elif data == "data3":
        # data3에 대한 처리 코드
        print("data3 처리 완료")
    else:
        # 나머지 데이터에 대한 처리 코드
        print("기타 데이터 처리")

# 큐가 비어있을 때까지 모든 데이터가 처리됩니다.

end_time = time.time()

execution_time = end_time - start_time
print(f"실행 시간: {execution_time} 초")
```

```python
import time
import queue

start_time = time.time()

# 큐 생성
my_queue = queue.Queue()

# 큐에 데이터 추가
for _ in range(50):
    my_queue.put("data1")
for _ in range(50):
    my_queue.put("data2")
for _ in range(50):
    my_queue.put("data3")

# 이곳에 수십 개의 데이터 추가

# 딕셔너리에 문자열 변수 매핑
data_mapping = {
    "data1": "data1_variable",
    "data2": "data2_variable",
    "data3": "data3_variable",
    # 나머지 데이터에 대한 변수 추가
}

# 변수 정의
data1_variable = "Data 1 Value"
data2_variable = "Data 2 Value"
data3_variable = "Data 3 Value"

# 함수를 딕셔너리에서 한 번 조회
process_data = lambda data: globals().get(data_mapping.get(data, None), None)

# 큐에서 데이터 분류 및 처리
while not my_queue.empty():
    data = my_queue.get()
    handler = process_data(data)
    if handler:
        print(f"{data} 처리 완료: {handler}")
    else:
        print(f"{data}에 대한 변수가 정의되지 않았습니다.")

# 큐가 비어있을 때까지 모든 데이터가 처리됩니다.

end_time = time.time()

execution_time = end_time - start_time
print(f"실행 시간: {execution_time} 초")
```
